```python
import pandas as pd

import matplotlib.pyplot as plt

# Read the CSV files
df1 = pd.read_csv('frame_log/yolo_detection.csv')
df2 = pd.read_csv('frame_log/florence_detection.csv')
df3 = pd.read_csv('frame_log/yolo_florence_detection.csv')
df4 = pd.read_csv('frame_log/baseline.csv')

# Plot the data
plt.figure(figsize=(10, 7))

plt.plot(df1['frame'], df1['person'], label='Person - YOLO')
plt.plot(df2['frame'], df2['person'], label='Person - Florence-2')
plt.plot(df3['frame'], df3['person'], label='Person - YOLO + Florence-2')
plt.plot(df4['frame'], df4['person'], label='Person - Baseline')

plt.plot(df1['frame'], df1['cat'], label='Cat - YOLO')
plt.plot(df2['frame'], df2['cat'], label='Cat - Florence-2')
plt.plot(df3['frame'], df3['cat'], label='Cat - YOLO + Florence-2')
plt.plot(df4['frame'], df4['cat'], label='Cat - Baseline')

plt.xlabel('Frame Index')
plt.ylabel('Count')
plt.title('Frame Log Data')
plt.legend(fontsize='small', loc='upper left')  # Set the font size of the legend labels to small and position it in the upper left corner
plt.show()
```
